```python
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline

im1 = plt.imread('sincan.jpg')
im2 = plt.imread('doraemon.jpg')

plt.imshow(im1, cmap='Greys_r')
```

```python
plt.imshow(im2, cmap='Greys_r')
```

```python
im1arr = np.array(im1)
im2arr = np.array(im2)
```

```python
import skimage
from skimage.transform import rescale
from skimage import img_as_float
```

```python
im1dp = skimage.img_as_float64(im1arr, force_copy=False)
im2dp = skimage.img_as_float64(im2arr, force_copy=False)
```

```python
imadd = np.add(im1, im2)
```

```python
# normalisasi
imaddNormal = imadd/(imadd.max())
```

```python
plt.imshow(imaddNormal, cmap='Greys_r')
```

```python
im1half = im1[0:200, 0:100, :]
im2half = im2[0:200,101:200, :]
```

```python
plt.imshow(im1half, cmap='Greys_r')
```

```python
plt.imshow(im2half, cmap='Greys_r')
```

```python
imhalfhalf = np.concatenate((im1half, im2half), axis =1)
plt.imshow(imhalfhalf, cmap='Greys_r')
```

```python
imLoop = np.empty((200,200,3))
for i in range(im1dp.shape[0]):
    for j in range(im1dp.shape[1]):
        for k in range(im1dp.shape[2]):
            if i%2==0:
                imLoop[i,:,:] = im1dp[i,:,:]
            else:
                imLoop[i,:,:] = im2dp[i,:,:]
```

```python
plt.imshow(imLoop, cmap='Greys_r')
```

```python
imNoLoop = np.empty((200,200,3))
imOdd=im1dp[::2]
imEven=im2dp[1::2]
imNoLoop[::2]=imOdd
imNoLoop[1::2]=imEven
```

```python
plt.imshow(imNoLoop, cmap='Greys_r')
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
gs = rgb2gray(imNoLoop)
```

```python
plt.imshow(gs, cmap='Greys_r')
title_obj = plt.title('Grayscale')
plt.setp(title_obj, color='black')  
```
